# Reinforcement Learning

Reinforcement Learning (RL) is a type of Machine Learning in which an **Agent** learns by interacting with an **Environment**.

Instead of learning from labeled data, the agent learns through **Trial and Error**.

The goal is to maximize the **Total Reward**.

---

# Real Life Example

Imagine a Robot trying to reach the treasure.

```

```

Every movement gives some reward.

```

Wall = -10

Goal = +100
```

Initially the robot knows nothing.

After many attempts, it learns the shortest path.

# Algorithm (Pseudo Code)

```
Initialize Q-table with zeros

Repeat for each Episode

    Select Current State

    Choose Action (ε-Greedy)

    Perform Action

    Observe Reward

    Observe Next State

    Update Q-value using

    Q(s,a)=Q(s,a)+α[R+γmaxQ(s',a')−Q(s,a)]

    Move to Next State

Repeat until Goal

Repeat for many Episodes
```

````python
import numpy as np
import gymnasium as gym
````

````python
env = gym.make("FrozenLake-v1", is_slippery=False)

print("States :", env.observation_space.n)
print("Actions :", env.action_space.n)
````

````python
q_table = np.zeros((16, 4))

print(q_table)
````

````python
#hyperparameters for rl
learning_rate = 0.8 #defining the right step to agent
gamma = 0.95 #how much you want your agent to explore 

epsilon = 1.0 #how much it explored 
epsilon_decay = 0.999
epsilon_min = 0.01

episodes = 20000 # 2000 times agent go from start to end 
````

````python
for episode in range(episodes):

    state, _ = env.reset()

    done = False

    while not done:

        # Epsilon-Greedy
        if np.random.uniform(0,1) < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(q_table[state])

        next_state, reward, terminated, truncated, _ = env.step(action)

        done = terminated or truncated

        # Q-Learning Formula
        q_table[state, action] = q_table[state, action] + learning_rate * (
            reward + gamma * np.max(q_table[next_state]) - q_table[state, action]
        )

        state = next_state

    if epsilon > epsilon_min:
        epsilon *= epsilon_decay
````

````python
print(q_table)
````

````python
state, _ = env.reset()

done = False

total_reward = 0

while not done:

    action = np.argmax(q_table[state])

    next_state, reward, terminated, truncated, _ = env.step(action)

    done = terminated or truncated

    total_reward += reward

    state = next_state

print("Reward :", total_reward)
````

````python
state = 0

best_action = np.argmax(q_table[state])

print("Best Action :", best_action)
````

# SARSA (State–Action–Reward–State–Action)

# Reinforcement Learning

SARSA is one of the most popular **On-Policy Reinforcement Learning Algorithms**.

It learns the best action by interacting with the environment and updating the Q-table based on the **actual next action chosen by the agent**.

Unlike Q-Learning, SARSA follows the same policy for both learning and action selection.

---

# Full Form of SARSA

```
S → State

A → Action

R → Reward

S → Next State

A → Next Action
```

This sequence gives the algorithm its name.

---

It learns by trying different paths repeatedly.

# SARSA Algorithm (Pseudo Code)

```
Initialize Q-table with zeros

For each Episode

    Select Current State

    Choose Action using ε-Greedy

    Repeat

        Perform Action

        Observe Reward

        Observe Next State

        Choose Next Action

        Update

        Q(s,a)=Q(s,a)+α[R+γQ(s',a')−Q(s,a)]

        State = Next State

        Action = Next Action

    Until Goal

Repeat for many Episodes
```

---

# Why is SARSA called On-Policy?

Because

```
Learning Policy

=

Behavior Policy
```

The algorithm learns from the **same action that it actually performs**, not from the best possible action.

---

# Q-Learning vs SARSA

| Q-Learning | SARSA |
|------------|--------|
| Uses Maximum Future Q-value | Uses Actual Next Action Q-value |
| Off-Policy | On-Policy |
| More Aggressive | More Conservative |
| Faster Learning | Safer Learning |
| May take risky shortcuts | Avoids risky actions |

````python
import numpy as np
import gymnasium as gym
````

````python
env = gym.make("FrozenLake-v1", is_slippery=False)

state_size = env.observation_space.n
action_size = env.action_space.n

print("States:", state_size)
print("Actions:", action_size)
````

````python
q_table = np.zeros((state_size, action_size))
````

````python
learning_rate = 0.8
gamma = 0.95

epsilon = 1.0
epsilon_decay = 0.999
epsilon_min = 0.01

episodes = 10000
````

````python
for episode in range(episodes):

    state, _ = env.reset()

    
    if np.random.rand() < epsilon:
        action = env.action_space.sample()
    else:
        action = np.argmax(q_table[state])

    done = False

    while not done:

        next_state, reward, terminated, truncated, _ = env.step(action)

        done = terminated or truncated

        # Choose next action (ε-greedy)
        if np.random.rand() < epsilon:
            next_action = env.action_space.sample()
        else:
            next_action = np.argmax(q_table[next_state])

       
        q_table[state, action] = q_table[state, action] + learning_rate * (
            reward +
            gamma * q_table[next_state, next_action] -
            q_table[state, action]
        )

        state = next_state
        action = next_action

    if epsilon > epsilon_min:
        epsilon *= epsilon_decay
````

````python
state, _ = env.reset()

done = False

total_reward = 0

while not done:

    action = np.argmax(q_table[state])

    next_state, reward, terminated, truncated, _ = env.step(action)

    done = terminated or truncated

    total_reward += reward

    state = next_state

print("Reward:", total_reward)
````
