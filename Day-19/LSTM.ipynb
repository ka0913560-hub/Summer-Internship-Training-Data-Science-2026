# Recurrent Neural Network (RNN)


---

# What is RNN?

Recurrent Neural Network (RNN) is a type of Deep Learning algorithm specially designed for **Sequential Data**.

Unlike ANN, RNN remembers previous information using a **Hidden State (Memory)**.

Examples of sequential data:

- Sentence
- Speech
- Stock Prices
- Weather Data
- Time Series
- DNA Sequence
- Sensor Data

---

# Why RNN?

Suppose you have a sentence:

```
I love Deep Learning
```

To understand the word **Learning**, the model must also remember:

```
I

в†“

love

в†“

Deep

в†“

Learning
```

ANN treats every word independently.

RNN remembers previous words.

---

# ANN vs RNN

| ANN | RNN |
|------|-----|
| No Memory | Has Memory |
| Independent Inputs | Sequential Inputs |
| Cannot remember previous input | Remembers previous inputs |
| Image Classification | NLP, Speech, Time Series |

# Hidden State Formula

\[
h_t = \tanh(W_x X_t + W_h h_{t-1} + b)
\]

Where

- \(X_t\) = Current Input
- \(h_{t-1}\) = Previous Hidden State
- \(W_x\) = Input Weight
- \(W_h\) = Hidden Weight
- \(b\) = Bias

---

# Output Formula

\[
Y_t = W_y h_t
\]

# What is BPTT?

Backpropagation Through Time is the training algorithm used in RNN.

Instead of updating weights for a single input, it updates them across all time steps.

# Problems in RNN

## Vanishing Gradient

Gradient becomes very small.

```
1

в†“

0.1

в†“

0.01

в†“

0.001
```

Network forgets old information.

---

## Exploding Gradient

Gradient becomes very large.

```
1

в†“

5

в†“

20

в†“

100
```

Training becomes unstable.

---

# Interview Questions

### What is RNN?

RNN is a neural network designed for sequential data that remembers previous information using hidden states.

---

### Why is RNN better than ANN?

Because RNN can use previous inputs while making the current prediction.

---

### What is Hidden State?

The hidden state is the memory of the RNN that carries information from previous time steps.

---

### What is BPTT?

Backpropagation Through Time is the algorithm used to train an RNN by propagating errors through all previous time steps.

---

### Why does RNN fail for long sequences?

Because of the **Vanishing Gradient Problem**, where gradients become extremely small and the network forgets earlier information.

# Key Takeaway

**ANN treats each input independently. RNN introduces a hidden state (memory) that carries information from previous time steps, making it suitable for sequence and time-series data. Its biggest limitation is the vanishing gradient problem, which is why LSTM and GRU were developed.**

````python
# Step 1: Import libraries
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
````

````python
# Step 2: Prepare a larger sample corpus (adding more varied sentences)
corpus = [
    "hello there how are you", "how are you doing today", "hello what are you up to",
    "how have you been", "hello what can i help you with", "how can i assist you",
    "good morning have a great day", "good night sleep tight", "hello how can I assist you",
    "what are you doing", "what can I help with today", "how can I make your day better"
]
````

````python

# Step 3: Tokenize the text into words
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
````

````python
# Convert words to sequences of integers
sequences = tokenizer.texts_to_sequences(corpus)

# Define sequence length and prepare input/output pairs for training
seq_length = 3
X = []
y = []
````

````python
for sequence in sequences:
    for i in range(seq_length, len(sequence)):
        X.append(sequence[i-seq_length:i])
        y.append(sequence[i])
````

````python
X = np.array(X)
y = np.array(y)
````

````python
# Convert labels to categorical
y = to_categorical(y, num_classes=len(tokenizer.word_index)+1)
````

````python
# Define Rnn Model
from tensorflow.keras.layers import SimpleRNN

model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=50, input_length=seq_length))
model.add(SimpleRNN(100, return_sequences=False))  # Changed from LSTM to SimpleRNN
model.add(Dropout(0.2))
model.add(Dense(len(tokenizer.word_index)+1, activation='softmax'))
````

````python
# Step 4: Define LSTM model with dropout for regularization
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=50, input_length=seq_length))  # Larger embedding size
model.add(LSTM(100, return_sequences=False))  # Using LSTM layer instead of SimpleRNN
model.add(Dropout(0.2))  # Regularization using dropout
model.add(Dense(len(tokenizer.word_index)+1, activation='softmax'))
````

````python

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
````

````python

model.fit(X, y, epochs=200, verbose=1)  # Increased epochs for better training
````

````python

# Step 6: Predict next word
def predict_next_word(sequence):
    sequence = tokenizer.texts_to_sequences([sequence])
    sequence = np.array(sequence)
    pred = model.predict(sequence, verbose=0)
    predicted_word_index = np.argmax(pred)
    predicted_word = tokenizer.index_word[predicted_word_index]
    return predicted_word
````

````python


# Test the model with an example input
input_sequence = "how can i"
predicted_word = predict_next_word(input_sequence)
print(f"Input: '{input_sequence}' в†’ Predicted next word: {predicted_word}")
````
