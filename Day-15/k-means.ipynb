   # K-Means Clustering Algorithm - Step by Step

## Step 1: Import Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

---

## Step 2: Load the Dataset

```python
data = pd.read_csv("data.csv")
print(data.head())
```

---

## Step 3: Select Features

```python
X = data.iloc[:, [0, 1]].values
```

---

## Step 4: Find the Optimal Number of Clusters (Elbow Method)

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

---

## Step 5: Create the K-Means Model

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
```

---

## Step 6: Train the Model

```python
kmeans.fit(X)
```

---

## Step 7: Predict Cluster Labels

```python
y_pred = kmeans.predict(X)
print(y_pred)
```

OR

```python
y_pred = kmeans.fit_predict(X)
print(y_pred)
```

---

## Step 8: Check Cluster Centers

```python
print(kmeans.cluster_centers_)
```

---

## Step 9: Visualize the Clusters

```python
plt.scatter(X[y_pred==0,0], X[y_pred==0,1], c='red', label='Cluster 1')
plt.scatter(X[y_pred==1,0], X[y_pred==1,1], c='blue', label='Cluster 2')
plt.scatter(X[y_pred==2,0], X[y_pred==2,1], c='green', label='Cluster 3')

plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            s=300,
            c='yellow',
            marker='X',
            label='Centroids')

plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

# Backend Working of K-Means

### Step 1
Choose the value of **K** (number of clusters).

Example:

```
K = 3
```

---

### Step 2
Randomly initialize **K Centroids**.

Example:

```
C1 = (2,3)
C2 = (7,8)
C3 = (9,2)
```

---

### Step 3
Calculate the distance of every data point from each centroid.

Euclidean Distance Formula

\[
Distance=\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}
\]

---

### Step 4
Assign each point to the nearest centroid.

```
Nearest Centroid → Cluster Assignment
```

---

### Step 5
Update the centroid by taking the mean of all points in that cluster.

Formula

\[
Centroid=\frac{\sum x}{n}
\]

---

### Step 6
Again calculate the distance of every point from the new centroids.

---

### Step 7
Repeat Steps 4–6 until centroids stop changing.

This is called **Convergence**.

---

## Final Output

- Cluster Labels
- Cluster Centers (Centroids)
- WCSS (Within Cluster Sum of Squares)
- Visualization of Clusters
