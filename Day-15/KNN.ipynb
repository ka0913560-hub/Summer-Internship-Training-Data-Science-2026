## K - Nearest Neighbor Algo:

### K-NN is a supervised Machine learning algorithm used for the classification tasks as well as regression task. It is simple and sensitive algorithm.


````python
import seaborn as sns
````

````python
df = sns.load_dataset('titanic')
````

````python
df.head()
````

````python
df = df[['survived' ,'pclass', 'sex' , 'age' , 'fare' , 'embarked']]
````

````python
df.head()
````

````python
df.isnull().sum()
````

````python
df.shape
````

````python
df.dropna(inplace = True)
````

````python
df.shape
````

````python
from sklearn.preprocessing import LabelEncoder , StandardScaler
````

````python
lb = LabelEncoder()
````

````python
df['sex'] = lb.fit_transform(df['sex'])
df['embarked'] = lb.fit_transform(df['embarked'])
````

````python
x = df.drop('survived' , axis = 1)
y = df['survived']
````

````python
sc = StandardScaler()
````

````python
x_scaled = sc.fit_transform(x)
````

````python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(x_scaled , y , test_size = 0.2 , random_state = 42)
````

````python
from sklearn.neighbors import KNeighborsClassifier
````

````python
clf = KNeighborsClassifier(n_neighbors=7)
````

````python
clf.fit(X_train , y_train)
````

````python
clf.score(X_test , y_test)
````

````python
pred = clf.predict(X_test)
````

````python
from sklearn.metrics import classification_report
````

````python
print(classification_report(y_test , pred))
````

````python
  from sklearn.ensemble import RandomForestClassifier
````

````python
clf1 = RandomForestClassifier(n_estimators=500)
clf1.fit(X_train, y_train)
````

````python
clf1.score(X_test , y_test)
````

````python
from sklearn.datasets import fetch_california_housing
````

````python
df1 = fetch_california_housing() 
````

````python
print(df1.DESCR)
````

````python
x = df1.data
y = df1.target
````

````python
from sklearn.preprocessing import StandardScaler
````

````python
sc = StandardScaler()
````

````python
x_scaled = sc.fit_transform(x)
````

````python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(x_scaled , y , test_size = 0.25 , random_state = 42)
````

````python
from sklearn.neighbors import KNeighborsRegressor
````

````python
model = KNeighborsRegressor(n_neighbors=15)
````

````python
model.fit(X_train , y_train)
````

````python
pred = model.predict(X_test)
````

````python
from sklearn.metrics import r2_score
````

````python
print(r2_score(y_test , pred))
````

````python
# Backend Working of KNN

## Step 1

Load the training dataset.

```
Training Data
```

↓

---

## Step 2

Choose the value of **K**.

Example

```
K = 3
```

or

```
K = 5
```

---

## Step 3

Take one new (unknown) data point.

Example

```
New Point = (Age = 30, Salary = 45000)
```

---

## Step 4

Calculate the distance from the new point to every training data point.

Most commonly used:

### Euclidean Distance

\[
Distance=\sqrt{\sum(x_i-y_i)^2}
\]

---

### Manhattan Distance

\[
Distance=\sum|x_i-y_i|
\]

---

## Step 5

Sort all distances from smallest to largest.

Example

```
Point A → 1.2

Point B → 2.5

Point C → 0.8

Point D → 3.1
```

↓

After Sorting

```
Point C

Point A

Point B

Point D
```

---

## Step 6

Pick the first **K nearest neighbors**.

Example

```
K = 3

Neighbor 1

Neighbor 2

Neighbor 3
```

---

## Step 7

Perform Majority Voting.

Example

```
Class A

Class A

Class B
```

↓

Majority

```
Class A
```

So prediction = **Class A**

---

## Step 8

Return the predicted class.

```
Prediction = Class A
```

---

# Complete Flow

Start

↓

Import Libraries

↓

Load Dataset

↓

Select Features & Target

↓

Split Dataset

↓

Feature Scaling

↓

Choose K

↓

Train KNN Model

↓

Calculate Distance

↓

Find K Nearest Neighbors

↓

Majority Voting

↓

Predict Class

↓

Evaluate Accuracy

↓

End
````
