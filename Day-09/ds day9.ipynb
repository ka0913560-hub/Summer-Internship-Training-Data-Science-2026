```python
import matplotlib.pyplot as plt
x=["mon","tue","wed","thur","fri"]
y=[12,4,12,13,14]

plt.title("sales-report")
plt.xlabel("days")
plt.ylabel("sales")
plt.plot(x,y,color="b")
plt.show()
```

```python
#pie chart

import matplotlib.pyplot as plt

subjects=["Python","SQL","POWER BI","Excel"]
students=[40,25,20,15]

plt.figure(figsize=(7,6))
plt.pie(
    students,
    labels=subjects,
    autopct="%1.1f%%",
    startangle=90,
    explode=(0.1,0,0,0)
)

plt.title("Students in Different Courses")
plt.show()
```

```python
#To make a hut

import matplotlib.pyplot as plt
from matplotlib.patches import Circle

x1=[2,2,4,4,2]
y1=[2,7,7,2,2]
plt.plot(x1,y1)

x2=[4,4,8,8,4]
y2=[2,7,7,2,2]
plt.plot(x2,y2)

x3=[2,3,4,2]
y3=[7,10,7,7]
plt.plot(x3,y3)

x4=[3,8,8,4,3]
y4=[10,10,7,7,10]
plt.plot(x4,y4)

x5=[2.5,2.5,3.5,3.5]
y5=[3,6,6,3]
plt.plot(x5,y5)

circle=Circle([3,8],0.5,fill=False)
plt.gca().add_patch(circle)

plt.savefig("Title.jpeg")
plt.show()
```

```python
import seaborn as sns
tip=sns.load_dataset("tips")
tip.head(6)
```

```python
iris=sns.load_dataset("iris")
iris.head(6)
```

```python
#1) line plot

sns.lineplot(x = 'tip' , y = 'total_bill' , data = tip)
plt.title('line-plot')
plt.show()
```

```python
#2) bar plot
sns.barplot(x = 'day' , y = 'total_bill' , data = tip)
```

```python
#3) # Countplot : It is used to count the occurences of categorical data.

sns.countplot(x='day' , data = tip)
```

```python
#4) boxplot - for outliers

sns.boxplot(x = 'day' , y = 'total_bill' , data = tip)
```

```python
#5) # violin plot:
#It is used for distribution and probability density.

sns.violinplot(x = 'day' , y = 'total_bill' , data = tip)
```

```python
#6) # swarm plot: 
#  It is used for display all the data points on categorical axis.

sns.swarmplot(x= 'day' , y = 'total_bill' , data = tip)
```

```python
#7) # pairplot:
#It is used to ploting pairwise relationships of dataset. 

sns.pairplot(iris , hue = 'species')
```

```python
#8) # heatmap :
# It is used to show the corelation of dataset.
coor = tip.corr()
cor = tip.corr(numeric_only = True)
sns.heatmap(cor , annot = True , cmap = 'coolwarm' )
```
